Welcome to the Linear Regression project! In this notebook, I will predict employee salaries based on their years of work experience. Linear regression allows us to forecast outcomes by finding the best-fit linear relationship in the data.The Mathematical ModelOur prediction follows the linear equation:$$y = wx + b$$Where:$w$ is the slope (weight)$b$ is the bias (intercept)$\epsilon$ (noise) represents the error due to real-world data imperfections.Optimization GoalTo find the optimal $w$ and $b$, we minimize the Residual Sum of Squares (RSS). This is also known as the loss function $L(w, b)$:$$L(w, b) = \sum_{i=1}^n (y_i - (wx_i + b))^2 \to \min$$Solving the SystemTo minimize $L(w, b)$, we take partial derivatives with respect to $w$ and $b$ and set them to zero:$$\frac{\partial L}{\partial w} = -2 \sum_{i=1}^n x_i (y_i - wx_i - b) = 0$$$$\frac{\partial L}{\partial b} = -2 \sum_{i=1}^n (y_i - wx_i - b) = 0$$This leads us to the Normal Equations system:
$$\begin{cases}
    \sum_{i=1}^n x_iy_i = w\sum_{i=1}^n x_i^2 + b\sum_{i=1}^n x_i \\
    \sum_{i=1}^n y_i = w \sum_{i=1}^nx_i + nb
\end{cases}$$ 

$$b = \bar{y} - w\bar{x}$$

$$w = \frac{n\sum xy - \sum x \sum y}{n\sum x^2 - (\sum x)^2}$$

Analytical solution $$\theta = (X^T X)^{-1} X^T y$$

```python
import math
import pandas as pd
dataset=pd.read_csv("Salary_Data.csv")
dataset
```

```python
import numpy as np

# 1. Загрузка данных (предполагаем, что dataset уже определен)
X = dataset["YearsExperience"].values
y = dataset["Salary"].values

# 2. Функции для среднего (лучше использовать встроенный np.mean, но сделаем вручную)
def get_mean(data):
    return np.sum(data) / len(data)

x_mean = get_mean(X)
y_mean = get_mean(y)

# 3. Расчет коэффициента w (Slope)
# Формула: sum((x - x_mean) * (y - y_mean)) / sum((x - x_mean)^2)
numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean)**2)
w = numerator / denominator

# 4. Расчет коэффициента b (Intercept)
# Формула: b = y_mean - w * x_mean
b = y_mean - (w * x_mean)

# 5. Вывод результатов
print(f"Коэффициент w (наклон): {w:.4f}")
print(f"Коэффициент b (смещение): {b:.4f}")

# Теперь можно сделать предсказание
y_pred = w * X + b
```

```python
import matplotlib.pyplot as plt

# 1. Создаем предсказания для всех точек X
# Это и есть наша прямая регрессии
Y_pred = w * X + b

# 2. Настройка графика
plt.figure(figsize=(10, 6)) # Размер окна


# 3. Рисуем реальные данные (Синие точки)
plt.scatter(X, y, color='royalblue', label='Реальные данные', alpha=0.7)

# 4. Рисуем линию регрессии (Красная линия)
plt.plot(X, Y_pred, color='crimson', linewidth=3, label='Модель линейной регрессии')

# 5. Оформление осей и заголовка
plt.title(f'Визуализация модели: Salary = {w:.1f} * Experience + {b:.1f}', fontsize=14)
plt.xlabel('Опыт работы (лет)', fontsize=12)
plt.ylabel('Зарплата', fontsize=12)
plt.legend() # Показать легенду

# 6. Отображение
plt.show()
```

```python
# 1. Предсказания для всех X
y_pred = w * X + b

# 2. Сумма квадратов остатков (ошибок)
ss_res = np.sum((y - y_pred)**2)

# 3. Общая сумма квадратов (отклонения от среднего)
y_mean = np.mean(y)
ss_tot = np.sum((y - y_mean)**2)

# 4. Расчет R^2
r_squared = 1 - (ss_res / ss_tot)

print(f"Коэффициент детерминации R^2: {r_squared:.4f}")
```
